```python
print("\n-----------------------------------")
print("Missing Values")
print("-----------------------------------")


missing_values = df.isnull().sum()
print(missing_values)
```

```python
#duplicate values

print("\n-----------------------------------")
print("Duplicate Records")
print("-----------------------------------")

duplicates = df.duplicated().sum()
print(f"Duplicate Rows: {duplicates}")
```

```python
print("\nPerforming basic cleaning...\n")

# Remove duplicate rows
df = df.drop_duplicates()

# ---------------------------------
# Fill Missing Numeric Values
# ---------------------------------

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

for column in numeric_columns:
    median_value = df[column].median()
    df[column] = df[column].fillna(median_value)

# ---------------------------------
# Fill Missing Categorical Values
# ---------------------------------

categorical_columns = df.select_dtypes(include=['object', 'string']).columns

for column in categorical_columns:
    mode_value = df[column].mode()[0]
    df[column] = df[column].fillna(mode_value)

print("Cleaning completed successfully.")
```

```python
import os

# Create folder for saving charts
os.makedirs("images", exist_ok=True)

sns.set_style("whitegrid")

# Histogram for Numeric Columns
numeric_cols = df.select_dtypes(include=np.number).columns

for column in numeric_cols[:3]:

    plt.figure(figsize=(8, 5))

    sns.histplot(df[column], kde=True)

    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")

    plt.tight_layout()

    plt.savefig(f"images/{column}_histogram.png")

    plt.close()
```

```python
# Correlation Heatmap
if len(numeric_cols) > 1:

    plt.figure(figsize=(10, 6))

    correlation = df[numeric_cols].corr()

    sns.heatmap(correlation,
                annot=True,
                cmap="coolwarm",
                fmt=".2f")

    plt.title("Correlation Heatmap")

    plt.tight_layout()
    plt.savefig("images/correlation_heatmap.png")
    plt.close()
```

```python
categorical_cols = df.select_dtypes(include='object').columns

if len(categorical_cols) > 0:

    top_column = categorical_cols[0]

    plt.figure(figsize=(10, 5))

    df[top_column].value_counts().head(10).plot(kind='bar')

    plt.title(f"Top Categories in {top_column}")
    plt.xlabel(top_column)
    plt.ylabel("Count")

    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig(f"images/{top_column}_barplot.png")
    plt.close()

# --------------------------------------------
# Final Message
# --------------------------------------------

print("\n-----------------------------------")
print("Task Completed Successfully")
print("-----------------------------------")

print("\nVisualizations are saved inside the images folder.")
```
